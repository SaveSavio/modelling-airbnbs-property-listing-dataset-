```python
import torch

# Assuming you have the actual and predicted values as tensors
actual_values = torch.tensor([1.0, 1.0, 1.0])
predicted_values = torch.tensor([1.0, 1.0, 2.0])

# Calculate the mean of actual values
mean_actual = torch.mean(actual_values)

# Calculate the total sum of squares
total_sum_squares = torch.sum((actual_values - mean_actual)**2)

# Calculate the sum of squared residuals
sum_squared_residuals = torch.sum((actual_values - predicted_values)**2)

# Calculate R^2
r_squared = 1 - (sum_squared_residuals / total_sum_squares)

print("Coefficient of Determination (R^2):", r_squared.item())
```

```python
import torch

labels = torch.tensor([1.0, 2.0, 1.0])
predictions = torch.tensor([1.0, 2.2, 3.8])

# Check if all labels are the same (to avoid division by zero)
if torch.all(torch.eq(labels, labels[0])):
    print("All labels are the same. R^2 is undefined.")
else:
    mean_labels = torch.mean(labels.float())
    
    # Ensure both tensors have the same data type
    labels = labels.float()
    predictions = predictions.float()

    sum_of_squared_residuals = torch.sum((labels - predictions)**2)  # SSR sum of squared residuals
    total_sum_of_squares = torch.sum((labels - mean_labels)**2)      # SST total sum of squares

    r2 = 1 - (sum_of_squared_residuals / total_sum_of_squares)

    print("Coefficient of Determination (R^2):", r2.item())
```
